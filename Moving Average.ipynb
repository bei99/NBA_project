```python
import pandas as pd
import numpy as np
```

```python
stats = pd.read_csv(r'NBA_scraping/data_cleaning/stats_merged.csv')
stats['GAME_DATE'] = pd.to_datetime(stats['GAME_DATE'])
stats.set_index(['GAME_DATE', 'HOME', 'AWAY'], inplace=True) 
stats = stats.drop('Unnamed: 0', 1)
stats.head(20)
```

```python
stats = stats.sort_index()
stats.head(20)
```

```python
weights = np.arange(1,11) #this creates an array with integers 1 to 10 included
stats_ma = stats.groupby('TEAM_ABBREVIATION').ewm(span=10, min_periods=10).mean()
stats_ma.head(20)
#stats_ma.to_csv(r'compare.csv')
stats_ma = stats_ma.shift(1)
stats_ma.head(15)
```

```python
stats_col = stats_ma.reset_index()
stats_home = stats_col
stats_home = stats_home[stats_home.TEAM_ABBREVIATION == stats_home.HOME]
stats_away = stats_col
stats_away = stats_away[stats_away.TEAM_ABBREVIATION == stats_away.AWAY]
```

```python
stats_home.set_index(['GAME_DATE', 'HOME', 'AWAY'], inplace=True)
stats_away.set_index(['GAME_DATE', 'HOME', 'AWAY'], inplace=True)
stats_home = stats_home.drop('TEAM_ABBREVIATION', 1)
stats_away = stats_away.drop('TEAM_ABBREVIATION', 1)

# Subtract stats_home - stats_away if they have the same index
stats_sub = stats_home - stats_away
stats_sub = stats_home - stats_away
stats_sub.head()
```

```python
stats_sub.to_csv(r'stats_ma_diff.csv')
```
